# Gather Tweets

```python
gossipPath = 'tweets'

filenames = []

for filename in os.listdir(gossipPath):
    print(filename)
    filenames.append(filename)
```

# Create Jobs for gathering replies

```python
interval = 200

# Create executor jobs for 5200 out of 5323 tweet files
# Each executor job contains executor jobs for 200 .csv files
for i in range(0, 26):
    
    commandLines = []
    start = i * interval
    end = (i+1) * interval
    print()
    print("************************************************")
    print("Interval from ", start, "to", end, "started.")
    print("************************************************")
    print()
    
    for filename in filenames[start:end]:
        print("Processing", filename)
        data = pd.read_csv(gossipPath+"/"+filename)
        for d in data['tweet_id']:
            commandLines.append("nasty replies --tweet-id "+ str(d) + " --to-executor jobs"+str(i)+".json")
        
    print("Creating executor job now...")
    for command in commandLines:
        !$command
```
